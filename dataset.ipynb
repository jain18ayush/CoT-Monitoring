```python
df = df[df['final_answer'] != '$']
```

```python
# drop the first two rows 
df = df.iloc[2:]
df
```

```python
# write to jsonl 
jsonl_path = "runs_cleaned.jsonl"  # Change to your file path
df.to_json(jsonl_path, orient='records', lines=True)
```
